```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.keys()
```

```python
iris_data = iris['data']
len(iris_data)
iris_data[:5]
```

```python
rotulos = iris['target_names']
rotulos
```

```python
classes = iris['target']
classes
```

```python
colunas = iris['feature_names']
colunas
```

```python
iris_df = pd.DataFrame(iris_data, columns=colunas)
iris_df['classes'] = classes
iris_df.head()
```

```python
iris_df.info()
```

```python
iris_df[iris_df.duplicated()]
```

```python
iris_df.drop_duplicates(inplace=True)
```

```python
iris_df.corr()
```

```python
import seaborn as sns
```

```python
sns.heatmap(iris_df.corr(), annot=True)
```

```python
iris_df.plot.box(vert=False)
```
